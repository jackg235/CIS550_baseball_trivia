```python
import numpy as np
import pandas as pd
import datetime
import requests, bs4
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import timedelta
```

```python
base_url = 'https://www.baseball-reference.com/postseason/'
end_url = '_NLCS.shtml'
browser = webdriver.Chrome(executable_path = '/Users/jackgoettle/Downloads/chromedriver')
```

```python
def to_int(num) :
    try :
        return int(num)
    except :
        return 0

def to_float(num) :
    try :
        return float(num)
    except :
        return 0

def scrape_batting(table, year, team) :
    ths = table.find_all('th')
    players = []
    count = 0
    for t in ths :
        if count < 31 :
            count += 1
            continue
        p = t.text
        p = p.replace('*', '')
        p = p.replace('#', '')
        if p != 'Totals' :
            d = {'NAME' : p}
            players.append(d)
    rows = table.find_all('tr')
    count = 0
    for r in rows :
        if count == len(players) + 2 :
            break
        if count < 2 :
            count += 1
            continue
        d = players[count - 2]
        d['YEAR'] = year
        d['TEAM'] = team
        tds = r.find_all('td')
        idx = 0
        for t in tds :
            if idx == 0 :
                d['G'] = to_int(t.text)
            elif idx ==1 :
                d['AB'] = to_int(t.text)
            elif idx ==2 :
                d['R'] = to_int(t.text)
            elif idx ==3 :
                d['H'] = to_int(t.text)
            elif idx ==4 :
                d['2B'] = to_int(t.text)
            elif idx ==5 :
                d['3B'] = to_int(t.text)
            elif idx ==6 :
                d['HR'] = to_int(t.text)
            elif idx ==7 :
                d['RBI'] = to_int(t.text)
            elif idx ==8 :
                d['BB'] = to_int(t.text)
            elif idx ==9 :
                d['SO'] = to_int(t.text)
            elif idx ==10 :
                d['BA'] = to_float(t.text)
            elif idx ==11 :
                d['OBP'] = to_float(t.text)
            elif idx ==12 :
                d['SLG'] = to_float(t.text)
            elif idx ==13 :
                d['OPS'] = to_float(t.text)
            elif idx ==14 :
                d['SB'] = to_int(t.text)
            elif idx ==15 :
                d['CS'] = to_int(t.text)
            elif idx ==16 :
                d['E'] = to_int(t.text)
            elif idx ==17 :
                d['WPA'] = to_float(t.text)
            elif idx ==18 :
                d['G SZN'] = to_int(t.text)
            elif idx ==19 :
                d['AB SZN'] = to_int(t.text)
            elif idx ==20 :
                d['R SZN'] = to_int(t.text)
            elif idx ==21 :
                d['H SZN'] = to_int(t.text)
            elif idx ==22 :
                d['HR SZN'] = to_int(t.text)
            elif idx ==23 :
                d['RBI SZN'] = to_int(t.text)
            elif idx ==24 :
                d['SB SZN'] = to_int(t.text)
            elif idx ==25 :
                d['BA SZN'] = to_float(t.text)
            elif idx ==26 :
                d['OPS SZN'] = to_float(t.text)
            idx += 1
        players[count - 2] = d
        count+=1
    return players
def scrape_pitching(table, year, team) :
    ths = table.find_all('th')
    players = []
    count = 0
    for t in ths :
        if count < 30 :
            count += 1
            continue
        p = t.text
        p = p.replace('*', '')
        p = p.replace('#', '')
        if p != 'Totals' :
            d = {'NAME' : p}
            players.append(d)
    rows = table.find_all('tr')
    count = 0
    for r in rows :
        if count == len(players) + 2 :
            break
        if count < 2 :
            count += 1
            continue
        d = players[count - 2]
        d['YEAR'] = year
        d['TEAM'] = team
        tds = r.find_all('td')
        idx = 0
        for t in tds :
            if idx == 0 :
                d['G'] = to_int(t.text)
            elif idx ==1 :
                d['GS'] = to_int(t.text)
            elif idx ==2 :
                d['ERA'] = to_float(t.text)
            elif idx ==3 :
                d['W'] = to_int(t.text)
            elif idx ==4 :
                d['L'] = to_int(t.text)
            elif idx ==5 :
                d['SV'] = to_int(t.text)
            elif idx ==6 :
                d['CG'] = to_int(t.text)
            elif idx ==7 :
                d['H'] = to_int(t.text)
            elif idx ==8 :
                d['R'] = to_int(t.text)
            elif idx ==9 :
                d['SO'] = to_int(t.text)
            elif idx ==10 :
                d['ER'] = to_float(t.text)
            elif idx ==11 :
                d['BB'] = to_int(t.text)
            elif idx ==12 :
                d['SO'] = to_int(t.text)
            elif idx ==13 :
                d['WHIP'] = to_float(t.text)
            elif idx ==14 :
                d['WPA'] = to_float(t.text)
            elif idx ==15 :
                d['G SZN'] = to_int(t.text)
            elif idx ==16 :
                d['GS SZN'] = to_int(t.text)
            elif idx ==17 :
                d['ERA SZN'] = to_float(t.text)
            elif idx ==18 :
                d['W SZN'] = to_int(t.text)
            elif idx ==19 :
                d['L SZN'] = to_int(t.text)
            elif idx ==20 :
                d['SV SZN'] = to_int(t.text)
            elif idx ==21 :
                d['IP SZN'] = to_float(t.text)
            elif idx ==22 :
                d['H SZN'] = to_int(t.text)
            elif idx ==23 :
                d['BB SZN'] = to_int(t.text)
            elif idx ==24 :
                d['SO SZN'] = to_int(t.text)
            elif idx ==25 :
                d['WHIP SZN'] = to_float(t.text)
            idx += 1
        players[count - 2] = d
        count+=1
    return players
```

```python
def scrape(year) :
    url = base_url + str(year) + end_url
    print(url)
    browser.get(url)
    browser.implicitly_wait(5)
    
    winner = browser.find_element_by_xpath("//*[@id='meta']/div/h1/span[2]")
    loser = browser.find_element_by_xpath("//*[@id='meta']/div/h1/span[3]")
    
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    table = soup.find_all("div", class_='table_outer_container')
    
    batting = scrape_batting(table[0], year, winner.text)
    batting = batting + scrape_batting(table[1], year, loser.text)
    
    pitching = scrape_pitching(table[2], year, winner.text)
    pitching = pitching + scrape_pitching(table[3], year, loser.text)
    return pd.DataFrame(batting), pd.DataFrame(pitching)
    
```

```python
batting = None
pitching = None
for yr in range (1969, 1980) :
    try : 
        b, p = scrape(yr)
        if batting is None :
            batting = b
        else :
            batting = batting.append(b)
        if pitching is None :
            pitching = p
        else :
            pitching = pitching.append(p)
    except :
        print('darn!')
        continue
```
